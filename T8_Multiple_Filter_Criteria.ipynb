# How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
# Applying a filter to the duration and sorting the result titles in alphabetical ascending order.
movies[movies['duration'] >= 200].sort_values('title')
```

**LOGICAL OPERATORS**

```python
# Or operator: 2 conditions and any of them evaluates to True, the result will be True.
print(True or False)
print(True or True)
print(False or False)
```

```python
# And operator: only True if both conditions are True.
print(True and False)
print(True and True)
print(False and False)
```

### MULTIPLE FILTERS

Now we want to specify another filter criteria. For example, we want to show only **long drama movies (200+ duration)**. 

```python
movies[(movies['duration'] >= 200) & (movies['genre']=='Drama')]
```

Now we want to specify another filter criteria. For example, we want to show only **long movies (200+ duration) or drama genre movies**. 

```python
movies[(movies['duration'] >= 200) | (movies['genre']=='Drama')]
```

Summary:

    - Each criteria inside parenthesis.
    
    - Criterias separated by logican operators.
    
That just gives a boolean pandas Series as we saw in the single criteria filter:
   * (movies['duration'] >= 200) | (movies['genre']=='Drama')

```python
type((movies['duration'] >= 200) | (movies['genre']=='Drama'))
```

```python
(movies['duration'] >= 200) | (movies['genre']=='Drama')
```

### BONUS TIP!

In the case we have a bunch of different "or" conditions for the same column (i.e: we want to show **movies which are either drama or crime or action**.

```python
movies[(movies['genre']=='Crime') | (movies['genre']=='Drama') | (movies['genre']=='Action')]
```

There is a way more efficient method to do the same:

```python
movies['genre'].isin(['Crime', 'Drama', 'Action'])
```

```python
movies[movies['genre'].isin(['Crime', 'Drama', 'Action'])]
```
